## Problem statement for FizzBuzz:

"Write a program that prints the numbers from 1 to 100. But for multiples of three print “Fizz” instead of the number and for the multiples of five print “Buzz”. For numbers that are multiples of both three and five print “FizzBuzz”."

For those of you more proficient in programming, you can solve the following for your interest. The point of the following exercise is to not only write accurate code, but also one that is readable and maintainable:

How would you scale the FizzBuzz problem to handle additional rules?
In addition to printing "Fizz" if the number is divisible by three, "Buzz" if the number is divisible by five, and "FizzBuzz" for both, you would have to incorporate additional rules: Print "Fizz" if the number is divisible by three, "Buzz" if it is divisible by 5, "Qizz" if it is divisible by 7, "FizzBuzz" for three and five, "FizzQizz" for three and seven, "BuzzQizz" for five and seven, and "FizzBuzzQizz" for three, five and seven. 

## Solution 1: Divisible by 3 and 5

````python
# First, we need to create the original list
hundred = list(range(1, 101))

# now, we can make an iterative process that checks each number in the list
#  based on its index value and substitute if necessary:

for i in range(0,100):
    num = hundred[i]

    # logic rules:
    if num % 3 == 0 and num % 5 == 0:
        hundred[i] = "FizzBuzz"
        print(hundred[i])
    elif num % 3 == 0 or num % 5 == 0:
        if num % 3 == 0:
            hundred[i] = "Fizz"
            print(hundred[i])
        elif num % 5 == 0:
            hundred[i] = "Buzz"
            print(hundred[i])
    else: 
        print(hundred[i])

# showing the final list
print(hundred)
    
````

## Solution 2: More Rules!

````python
# First, we need to recreate the original list since it was overwritten with strings
hundred = list(range(1, 101))

# now, we can make an iterative process that checks each number in the list
#  based on its index value and substitute if necessary:

for i in range(0,100):
    num = hundred[i]

    # logic rules:
    if num % 3 == 0:
        if num % 5 == 0:
            if num % 7 == 0:
                hundred[i] = "FizzBuzzQizz"
                print(hundred[i])
            else:
                hundred[i] = "FizzBuzz"
                print(hundred[i])
        elif num % 7 == 0:
            hundred[i] = "FizzQizz"
            print(hundred[i])
        elif num % 7 != 0 and num % 5 != 0: 
            hundred[i] = "Fizz"
            print(hundred[i])
    elif num % 5 == 0:
        if num % 7 == 0 and num % 3 != 0:
            hundred[i] = "BuzzQizz"
            print(hundred[i])
        elif num % 7 != 0 and num % 3 != 0:
            hundred[i] = "Buzz"
            print(hundred[i])
    elif num % 7 == 0 and num % 3 != 0 and num % 5 != 0:
        hundred[i] = "Qizz"
        print(hundred[i])
    else: 
        print(hundred[i])

# showing the final list
print(hundred)
````

## Problem Statement for Tic Tac Toe:
Given an existing board of tic tac toe, find the winner of the game.

Sample input into your function: 

``` 
board = [['O', 'X', 'X'],
         ['X', 'X', 'O'],
         ['O', 'X', 'X']] 
```
         
Sample output from your function: 'X won', or 'O won', or 'Draw'.

````python
# the first step would be to have the user input the values they are testing:

a1 = input("What is the value in A1? Please use 'X' or '0' for the set values, and any other letter for spaces that did not get used.")
a2 = input("What is the value in A2? Please use 'X' or '0' for the set values, and any other letter for spaces that did not get used.")
a3 = input("What is the value in A3? Please use 'X' or '0' for the set values, and any other letter for spaces that did not get used.")
b1 = input("What is the value in B1? Please use 'X' or '0' for the set values, and any other letter for spaces that did not get used.")
b2 = input("What is the value in B2? Please use 'X' or '0' for the set values, and any other letter for spaces that did not get used.")
b3 = input("What is the value in B3? Please use 'X' or '0' for the set values, and any other letter for spaces that did not get used.")
c1 = input("What is the value in C1? Please use 'X' or '0' for the set values, and any other letter for spaces that did not get used.")
c2 = input("What is the value in C2? Please use 'X' or '0' for the set values, and any other letter for spaces that did not get used.")
c3 = input("What is the value in C3? Please use 'X' or '0' for the set values, and any other letter for spaces that did not get used.")

board = [[a1, a2, a3], [b1, b2, b3], [c1, c2, c3]]

print(board)

x_won = ["X", "X", "X"]
o_won = ["O", "O", "O"]

# for checking the lists, I just made the X-based list combinations that could possibly win
# then just copied it to O

if [board[0][0], board[0][1], board[0][2]] == x_won or \
    [board[1][0], board[1][1], board[1][2]] == x_won or \
    [board[2][0], board[2][1], board[2][2]] == x_won or \
    [board[0][0], board[1][0], board[2][0]] == x_won or \
    [board[0][1], board[1][1], board[2][1]] == x_won or \
    [board[0][2], board[1][2], board[2][2]] == x_won or \
    [board[0][0], board[1][1], board[2][2]] == x_won or \
    [board[0][2], board[1][1], board[2][0]] == x_won:
    
    print("X Won")

elif [board[0][0], board[0][1], board[0][2]] == o_won or \
    [board[1][0], board[1][1], board[1][2]] == o_won or \
    [board[2][0], board[2][1], board[2][2]] == o_won or \
    [board[0][0], board[1][0], board[2][0]] == o_won or \
    [board[0][1], board[1][1], board[2][1]] == o_won or \
    [board[0][2], board[1][2], board[2][2]] == o_won or \
    [board[0][0], board[1][1], board[2][2]] == o_won or \
    [board[0][2], board[1][1], board[2][0]] == o_won:
    
    print("O Won")

else: 
    print("Draw")
````
